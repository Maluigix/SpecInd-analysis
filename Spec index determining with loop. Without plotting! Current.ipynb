```python
def rSDSSFITS(FITSfile):
    with fits.open(FITSfile) as hdulist:
        h0=hdulist[0].header
        d0=hdulist[0].data
        h1=hdulist[1].header
        d1=hdulist[1].data
        d2=hdulist[2].data
        h2=hdulist[2].header
        
        n=d1.shape[-1]
        lw=d1['loglam']
        z=d2['Z']
        zer=d2['Z_ERR']
        RA=h0['RA']
        mjdate=h0['MJD']
        platte=h0['PLATEID']
        fibb=h0['FIBERID']
        DEC=h0['DEC']
        wl=np.power(10.0,lw)
        ivar=d1['ivar']
        #[-np.inf if x<=0 else x for x in ivar]
        #sig=np.reciprocal(np.sqrt(ivar)) 
        #'sig':sig,
        #er=ivar
        # incorrect and fix!!! 
      
    return {'wl':wl,'mjdate':mjdate,'platte':platte,'fibb':fibb,'lw':lw,'fl':d1['flux'],'zer':zer,'DEC':DEC,'RA':RA,'er':d1[2],'ivar':ivar,'h0':h0,'d0':d0,'d1':d1,'h1':h1,'h2':h2,'d2':d2,'z':z,'zer':zer}
```

```python
def func_powerlaw(lam, m, c):
    return c*lam**m
#DLAzval=[]
#DLAspecival=[]
#DLAraval=[]
#DLAdecva=[]
spre=np.array([[1276,1292],[1314,1328],[1345,1362],[1435,1465],[1684,1700]])
for i in range(0,4868):
    if i%100==0:
        print(i)
#qc
#DLAQUANAMES
    FITSfile='spe/'+DLAQUANAMES[i]
    s=rSDSSFITS(FITSfile)
    
    x=s['wl']
    zerr=s['zer']
    decli=s['DEC']
    riasc=s['RA']
    z=s['z']
    y=np.array(s['fl'])
    z1=z+1
    ospre=spre*z1
    medspreg=[]
    for i in range(0,5):
            g0=np.median(ospre[i])
            medspreg.append(g0)
    fval=[]
    for i in range(0,5):
            g1=y[(x>=ospre[i][0]) & (x<=ospre[i][1])]
            fval.append(g1)
    iqrval=[]
    for i in range(0,5):
            g2=iqr(fval[i],rng=(25,75))
            iqrval.append(g2)
    fmed=[]
    for i in range(0,5):
            g3=np.median(fval[i])
            fmed.append(g3)
    X=medspreg
    Y=fmed
    popt, pcov = curve_fit(func_powerlaw,X,Y,sigma=iqrval,maxfev=130000)
    DLAspecival.append(popt[0])
    DLAzval.append(z)
    DLAraval.append(riasc)
    DLAdecva.append(decli)
```
